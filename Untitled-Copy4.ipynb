```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from math import sin, sqrt, atan, cos
from sympy.abc import z
from IPython.display import display
from sympy.functions.elementary.complexes import arg
from sympy.functions.elementary.complexes import re


#   
mu = 1 # 
w = 1 # 
phi = 1 # 
delta_t = 0.25 # 
n = 100 #   ( )
x = [0]

#   :
# 1 : 1+sin(1*x+1)
# 2 : 1+sin(1*x+0) + sin(0.5*x+0)
# 3 : 1+sin(x)+sin(2*x)+sin(3*x)


func = sp.sympify("1+sin(x+1)+sin(2*x+1)")

for i in range(n):
    x.append(x[i]+delta_t)

y = [func.subs(sp.Symbol('x'), x[i]) for i in range(len(x))]

plt.plot(x, y, label="")
plt.legend()
plt.grid()
plt.show()

matrix = np.zeros((n+1,n+1)) #   101  101  

matrix[0][0] = 1 #     

for i in range(n):
    matrix[1][i] = y[i] #     
eps = 10**-3

for l in range(2,n):
    for k in range(0,n+1-l):
        matrix[l][k] = (matrix[l-2][k+1]/matrix[l-2][0] - matrix[l-1][k+1]/matrix[l-1][0])
    print(sum(abs(matrix[l])))
    if sum(abs(matrix[l])) < eps: #     
        print("l=",l)
        m = l + 1 #   
        break
print("m =",m)
print("n =",(m-4)/4) #  
n = int((m-4)/4) #      
drob = str(y[0]) + "/" #        

for i in range(2,m-1): #  
    drob += "(1+("
    drob += str(matrix[i][0]) + "*z**-1)"
    drob += "/"
drob = drob[:-1] #    "/" 
#print(drob) 
for i in range(2,m-1):
    drob += ")" #  m-1  
c = sp.sympify(drob, evaluate=False)
display(c) #  
a = sp.N(sp.simplify(drob)) #    
display(a)

#     

b = str(sp.simplify(a))
index = b.find('/') + 1
znam = b[index:]
print(znam)
```

```python
roots = sp.solve(znam) #  solve  Sympy     
print(roots)
```

```python
type_complex = "5 + 2*I" 
type_complex = sp.simplify(type_complex) 
#    ,        

eq_args = []
for i in range(len(roots)):
    if type(roots[i])==type(type_complex): #    
        eq_args.append(arg(roots[i])) #   ,      arg 
print(eq_args)
```

```python
wi = [] 
eq_args.sort()
eq_args_new = []

#     
for i in range(len(eq_args)):
    if eq_args[i] > 0:
        eq_args_new.append(eq_args[i])

#    
for i in range(n):
    wi.append(eq_args_new[i]/delta_t)

print(wi)
```

```python
#         ,   mu+ bi=  
#       ,            
first_eq = "mu+"
for i in range(n):
    first_eq += "b" + str(i) + "+"
first_eq = first_eq[:-1]
first_eq += "==" + str(y[0])


#     2*n+1 ,     2*n+1 .
#    ,  2*n
equations = []
for i in range(2*n):
    y_in_delta = func.subs(sp.Symbol('x'), (i+1)*delta_t) #     
    our_sum = "mu+" #     
    for j in range(n):
        our_sum += "a" + str(j) + "*sin(" + str(i+1) + "*" + str(wi[j]) + "*" + str(delta_t) + ")+b" + str(j) + "*cos(" + str(i+1) + "*" + str(wi[j]) + "*" + str(delta_t) + ")+"
    our_sum = our_sum[:-1] #     
    our_sum += "==" + str(y_in_delta) #  .   
    equations.append(our_sum)
for i in range(2*n):
    print(equations[i]) 
first_eq = sp.N(sp.sympify(first_eq,evaluate=False)) 
equations = [sp.N(sp.sympify(equations[i], evaluate=False)) for i in range(len(equations))]
equations.append(first_eq) #     
print("System:")
for i in range(len(equations)):
    print(equations[i])
unknowns = []

#    .   2*n+1
#    
#   a0,a1,...,an b0,b1,...,bn
#          
# a0,a1,...,an,b0,b1,...bn,mu 
for i in range(n):
    unknowns.append(sp.Symbol('a'+str(i)))
for i in range(n):
    unknowns.append(sp.Symbol('b'+str(i)))
unknowns.append(sp.Symbol('mu'))
res = sp.linsolve(equations, unknowns) #   equations,      (unknowns)
print("Result:", res)

#    :
# a0,a1,...,an,b0,b1,...bn,mu 
#      a0,b0  : [0][0], [0][0+n]
#                      a1,b1  : [0][1], [0][1+n]
#   
#     (2*n),  +1, ..    
c = [sqrt(pow(res.args[0][i],2) + pow(res.args[0][i+n], 2)) for i in range(n)]
phi = [atan(res.args[0][i+n]/res.args[0][i]) for i in range(n)]
a = [c[i]*cos(phi[i]) for i in range(n)]
b = [c[i]*sin(phi[i]) for i in range(n)]
mu = res.args[0][2*n]

harmonics = []
y_harmonics = []
for i in range(n):
    #   
    #    n .
    harmonics.append(sp.sympify(str(mu) + "+" + str(a[i]) + "*sin(" + str(wi[i]) + "*x)+" + str(b[i]) + "*cos(" + str(wi[i]) + "*" + "x)"))
    y_harmonics.append([])
    y_harmonics[i] = [harmonics[i].subs(sp.Symbol('x'), x[j]) for j in range(len(x))] # 
print("Harmonics:")

for i in range(n):
    print("Harmonic " + str(i) + ":", end="")
    print(harmonics[i])

for i in range(n):
    plt.plot(x, y_harmonics[i], label=str("harmonic " + str(i+1)))
plt.legend()
plt.grid()
plt.show()
```

```python
#                   
y_polyharmony = []
for i in range(len(x)):
    y_polyharmony.append(0)
for i in range(len(x)):
    for j in range(n):
        y_polyharmony[i] += y_harmonics[j][i]
    y_polyharmony[i] -= mu*(n-1) #         
plt.plot(x, y_polyharmony, label='Polyharmony')
plt.plot(x, y, label='Original Function')
plt.grid()
plt.legend()
plt.show()
```
